# Starting with Code :

### Importing Libraries :

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import nltk
import math
import re
```

```python
df = pd.read_csv('IMDB Dataset.csv')
```

```python
df.head()
```

## Text Preprocessing :

### Convert text to lower case :

```python
df['review'] = [str(i).lower() for i in df['review']]
    
```

```python
df.head()
```

### Tokenisation :

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
```

```python
sentence_tokenize = []
for sentence in df['review']:
    sentence = sent_tokenize(sentence)
    sentence_tokenize.append(sentence)
```

```python
sentence_tokenize
```

### Word Tokenizer :

```python
word_tokenize = [word_tokenize(word) for word in df['review']]
```

### Cleaning Dataset :

```python
import re

lst = []

for words in word_tokenize:
    clean = []
    for word in words:
        res = re.sub(r'[^\w\s]',"", word)
        if res != "":
            clean.append(res)
        lst.append(clean)
```
